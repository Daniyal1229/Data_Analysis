```python
import pandas as pd 
import matplotlib
import statistics as sta
#loading DataFrame
data= pd.read_csv('Zombaie.csv')
num_rows = data.shape[0]
num_cols = data.shape[1]

# Print the result
print("The data set has", num_rows, "rows and", num_cols, "columns.")
# print(data)
data.head(20)
```

```python


# Calculate the total number of records in the data set
total_records = len(data)

# Calculate the number of records that have been impacted by the zombie virus
impacted_records = len(data[data['Zombies'] == True])

# Calculate the percentage of records that have been impacted by the zombie virus
percentage_impacted = round((impacted_records / total_records) * 100)

# Print the percentage of records that have been impacted by the zombie virus
print(f"{percentage_impacted}% of the data set has been impacted by the zombie virus.")
```

```python
# Calculate the total number of records in the data set
total_records = len(data)

# Calculate the number of records that have been impacted by the zombie virus
impacted_records = len(data[data['Zombies'] == True])

# Calculate the number of impacted individuals for each gender
male_impacted = len(data[(data['Zombies'] == True) & (data['gender'] == 'Male')])
female_impacted = len(data[(data['Zombies'] == True) & (data['gender'] == 'Female')])

# Calculate the percentage of impacted individuals for each gender
male_percentage_impacted = round((male_impacted / impacted_records) * 100)
female_percentage_impacted = round((female_impacted / impacted_records) * 100)

# Print the percentage of impacted individuals for each gender
print(f"{male_percentage_impacted}% of impacted individuals were male, and {female_percentage_impacted}% were female.")
```

```python
import numpy as np
# Calculate the mean age
mean_age = round(np.mean(data['age']))

# Calculate the mode age
mode_age = round(data['age'].mode()[0])

# Calculate the median age
median_age = round(np.median(data['age']))

# Calculate the range of ages
min_age = min(data['age'])
max_age = max(data['age'])
age_range = max_age - min_age

# Print the results
print(f"Mean age: {mean_age}\nMode age: {mode_age}\nMedian age: {median_age}\nAge range: {age_range}")
```

```python
infected_mean_age = round(np.mean(data[data['Zombies'] == True]['age']))

# Calculate the mean age of the uninfected population
uninfected_mean_age = round(np.mean(data[data['Zombies'] == False]['age']))

# Print the results
print(f"Mean age of infected population: {infected_mean_age}\nMean age of uninfected population: {uninfected_mean_age}")
```

```python


# Group the data by state and sum the number of zombies in each state
zombie_counts = data.groupby('location')['Zombies'].sum()

# Sort the data in descending order by the number of zombies
zombie_counts = zombie_counts.sort_values(ascending=False)

# Extract the state with the third highest number of zombies
third_highest_state = zombie_counts.index[2]

# Print the result
print(f"The state with the third highest number of zombies is {third_highest_state}.")
```
